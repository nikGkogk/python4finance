```python
import numpy as np
import pandas as pd

filename = '../data/tr_eikon_eod_data.csv'
f = open(filename, 'r')
f.readlines()[:5]  # Show raw data
```

```python
data = pd.read_csv(filename, index_col='Date') # make sure to define index column
data.index = pd.to_datetime(data.index) # convert the index to actual date objects
data.info() # Show dataFrame infor
```

```python
data.head() # first rows as dataFrame, .tail() for last rows
```

```python
data.tail()
```

```python
instruments = ['Apple Stock', 'Microsoft Stock',
'Intel Stock', 'Amazon Stock', 'Goldman Sachs Stock',
'SPDR S&P 500 ETF Trust', 'S&P 500 Index',
'VIX Volatility Index', 'EUR/USD Exchange Rate',
'Gold Price', 'VanEck Vectors Gold Miners ETF',
'SPDR Gold Trust']
for ric, name in zip(data.columns, instruments):
    print('{:8s} | {}'.format(ric, name))
```

```python
data.describe().round(2) # useful standard statistics per column/instrument
```

```python
data.mean(numeric_only=True) # mean value per column
```

```python
data.agg(['min', 'mean', 'std', 'median', 'max']) # customize the aggregate information
```

```python
data.diff().head() # provides the absolute changes between two index values.
```

```python
data.diff().mean() # aggregation operations can be applied
```

```python
data.pct_change(fill_method=None).round(3).head() # percentage of change
```

```python
data.pct_change(fill_method=None).mean().plot(kind='bar', figsize=(10, 6));
```

```python
rets = np.log(data / data.shift(1)) # log returns in vectorised fashion
rets.cumsum().apply(np.exp).plot(figsize=(10, 6));
```

```python
data.resample('1W', label='right').last().head() # resampling to weekly data
```

```python
# This plots the cumulative log returns over time: first, the cumsum() method is
# called, then np.exp() is applied to the results; finally, the resampling takes place.
rets.cumsum().apply(np.exp).resample('1ME', label='right').last().plot(figsize=(10, 6));
```

```python
rets.cumsum().apply(np.exp).resample('1ME', label='right').last().plot(figsize=(10, 6));
```
